```python
dat = np.loadtxt('G3.points')

res = np.loadtxt('chains_G3_sys/G3_res.points')
mod = np.loadtxt('chains_G3_sys/G3_mod.points')
```

```python
plt.subplot(221)
plt.errorbar(dat[:,0],dat[:,1],yerr=res[:,3],ls='None')
plt.errorbar(dat[:,0],dat[:,1],yerr=dat[:,3],ls='None',color='red')
plt.plot(mod[:,0],mod[:,2])
plt.ylabel('X [as]')

    
plt.subplot(222)
plt.errorbar(res[:,0],res[:,1],yerr=res[:,3],ls='None')
plt.plot(mod[:,0],mod[:,1]-mod[:,2],'g')
plt.plot(mod[:,0],mod[:,3]-mod[:,2],'g')
plt.ylabel('Residuals X [mas]')
    
plt.subplot(223)
plt.errorbar(dat[:,0],dat[:,2],yerr=dat[:,4],ls='None')
plt.errorbar(dat[:,0],dat[:,2],yerr=dat[:,4],ls='None',color='red')
plt.plot(mod[:,0],mod[:,5])
plt.ylabel('Y [as]')
    
plt.subplot(224)
plt.errorbar(res[:,0],res[:,2],yerr=res[:,4],ls='None')
plt.plot(mod[:,0],mod[:,4]-mod[:,5],'g')
plt.plot(mod[:,0],mod[:,6]-mod[:,5],'g')
plt.ylabel('Residuals Y [mas]')
    
plt.gcf().set_size_inches((10,10))
plt.tight_layout()
plt.show()
```

```python
plt.errorbar(dat[:,1],dat[:,2],xerr=dat[:,3],yerr=dat[:,4],ls='None')
plt.plot(mod[:,2],mod[:,5])

plt.axis('equal')

plt.xlim([-0.1,.9])
plt.ylim([-.4,.3])
plt.show()
```

```python
datR = np.loadtxt('G3.rv')

resR = np.loadtxt('chains_G3_sys/G3_res.rv')
modR = np.loadtxt('chains_G3_sys/G3_mod.rv')
```
